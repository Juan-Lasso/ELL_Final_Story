```python
## import pandas 
import pandas as pd 
```

```python
## let's read the file 
df = pd.read_excel("raw-data/Demographic_Snapshot_ENL.xlsx", sheet_name = "School")
df.info()
```

```python
## let's eliminate unwanted columns 
df_clean = df[["School Name", "Year", "Total Enrollment", "# English Language Learners", "% English Language Learners"]]
df_clean
```

```python
df_clean['Start Year'] = df_clean['Year'].str.extract(r'(\d{4})').astype(int)
df_clean
```

```python
school_ell_trends = df_clean.pivot_table(
    index='School Name',
    columns='Start Year',
    values='# English Language Learners'
)
school_ell_trends
```

```python
school_ell_trends.head()
```

```python
# How many unique schools are in the filtered DataFrame?
original_school_count = df_clean['School Name'].nunique()
original_school_count
```

```python
# How many schools made it into the pivot table?
pivot_school_count = school_ell_trends.shape[0]
pivot_school_count
```

```python
## Let's replace the NaN values with 0 
school_ell_trends = school_ell_trends.fillna(0)
school_ell_trends
```

```python
# 1. Percentage change between 2020 and 2021
school_ell_trends['Change_2020_2021'] = ((school_ell_trends[2021] - school_ell_trends[2020]) / school_ell_trends[2020]) * 100

# 2. Percentage change between 2019 and 2021
school_ell_trends['Change_2019_2021'] = ((school_ell_trends[2021] - school_ell_trends[2019]) / school_ell_trends[2019]) * 100

# 3. Percentage change between 2021 and 2023
school_ell_trends['Change_2021_2023'] = ((school_ell_trends[2023] - school_ell_trends[2021]) / school_ell_trends[2021]) * 100
school_ell_trends
```
